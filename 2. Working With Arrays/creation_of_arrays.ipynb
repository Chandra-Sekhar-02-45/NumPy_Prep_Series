## Creating NumPy Arrays (ndarray)

NumPy arrays are the foundation of numerical computing in Python. Let's learn different ways to create them!

```python
import numpy as np
```

### 1. Creating Arrays from Python Objects

You can create NumPy arrays from common Python data structures like **lists** and **tuples**.

Let's see how this works with examples:

**Example 1: Creating array from a Python List**

```python
list_a = [1 , 2 ,3, 4] 
numpy_array = np.array(list_a)
numpy_array
```

**Example 2: Creating array from a Python Tuple**

```python
tuple_a = (7 , 2 ,3, 4) 
numpy_array_from_tuple = np.array(tuple_a)
numpy_array_from_tuple
```

**Let's check the type of both arrays:**

```python
print(type(numpy_array))
print(type(numpy_array_from_tuple))
```

### Key Takeaways:

* **Same Output Type**: Both list and tuple inputs create the 
same type â†’ `numpy.ndarray`

* **Automatic Type Detection**: NumPy automatically figures out the appropriate data type for your array

* **Homogeneous Data**: All elements in a NumPy array are converted to the same data type (dtype)

* **Syntax**: Simply use `np.array()` and pass your list or tuple inside the parentheses
